```python
import pandas as pnds
#dependency for pandas with alias

import os
#using pathing style regardless of os
```

```python
# gets indirect path to a file, os.path.join() to load a file from somewhere in directory
dload_school = os.path.join(".","Resources", "schools_complete.csv")
dload_students = os.path.join(".", "Resources", "students_complete.csv")

print(dload_school)
print(dload_students)
```

```python
#reading csv files thro dataframe "df"

school_data_df = pnds.read_csv(dload_school)

school_data_df.head(3)
#views the first 3 rows

school_data_df.tail(4)
#views the last 4 rows

school_count = len(school_data_df['school_name'].unique())

school_count
```

```python
# Read the student data file and store it in a Pandas DataFrame.
student_data_df = pnds.read_csv(dload_students)
student_data_df.head(5)
```

```python
#checking nullvalues (filled values) using count, isnull, notnull
#true if null, false if filled
school_data_df.count()

student_data_df.count()

school_data_df.isnull()

student_data_df.isnull()

#count up all the null (empty) values
student_data_df.isnull().sum()

#notnull returns true for filled, false for empty, opposite of isnull
student_data_df.notnull().sum()
```

```python
# Add each prefix and suffix to remove to a list.
prefixes_suffixes = ["Dr. ", "Mr. ","Ms. ", "Mrs. ", "Miss ", " MD", " DDS", " DVM", " PhD"]

for joke in prefixes_suffixes:
    student_data_df['student_name'] = student_data_df['student_name'].str.replace(joke, "")
    
student_data_df.head(15)
    
```

```python
# merging both student_data_df and school_data_df with a shared school_name coloumn

cmplt_stuskool_df = pnds.merge(student_data_df, school_data_df, on = ['school_name', 'school_name'])

cmplt_stuskool_df.head(10)

# ^ merge(a_df, b_ad, on = ['c','c']) prioritizes a_df on the left side, b_df on the right

# cmplt_stuskool_df = pnds.merge(school_data_df, student_data_df, on = ['school_name', 'school_name'])

# cmplt_stuskool_df.head(10)

# while ^ priotizes b_df on the left, a_df on the right

#make sure matching coloumn names are the same, if not, change coloumn name
```

```python
#counting students incase missing

# count_studs = cmplt_stuskool_df.count()

# count_studs 

count_studs = cmplt_stuskool_df["Student ID"].count()
count_studs
```

```python
# unique() to count all the unique values

len(cmplt_stuskool_df["school_name"].unique())
```

```python
# total budget of all the schools

ttl_budget = sum(cmplt_stuskool_df["budget"].unique())

ttl_budget
```

```python
# avaerage (mean) of reading and math scores
avg_read_score = cmplt_stuskool_df["reading_score"].mean()

avg_read_score

avg_math_score = cmplt_stuskool_df["math_score"].mean()

avg_math_score
```

```python
#percentage of students getting 70 or better in reading , math
#overall percentage as well

mth_passing = cmplt_stuskool_df['math_score']>=70
mth_passing

#  returns bool values, so filter out false, and keep true (>70)

mth_passing = cmplt_stuskool_df[cmplt_stuskool_df["math_score"]>=70]

mth_passing.head()


read_passing = cmplt_stuskool_df['reading_score']>=70
read_passing

#  returns bool values, so filter out false, and keep true (>70)

read_passing = cmplt_stuskool_df[cmplt_stuskool_df["reading_score"]>=70]

read_passing.head()
```

```python
# find total number of passing math, reading scores

# len(mth_passing)

# len(read_passing)

#why not len?

math_passed = mth_passing["student_name"].count()

reading_passed = read_passing["student_name"].count()
```

```python
# pencentage of math reading passed out of total student count
# total_math_passed = ((mth_passing['student_name'].count() / count_studs)*100)

# total_math_passed

# total_reading_passed = (((read_passing['student_name'].count()/ count_studs)*100))

# total_reading_passed

math_passed_perc = ((math_passed / count_studs)*100)

math_passed_perc

reading_passed_perc = (((reading_passed / count_studs)*100))

reading_passed_perc
```

```python
# number of studends who passed both math and reading 

both_passed = cmplt_stuskool_df[(cmplt_stuskool_df["math_score"]>=70) & (cmplt_stuskool_df["reading_score"]>=70)]

both_passed['student_name'].count()

both_passed_count = both_passed['student_name'].count()

both_passed_count

# percentage of students who passed both math and reading

both_passed_perc = ((both_passed['student_name'].count()/ count_studs)* 100)

both_passed_perc
```

```python
# Adding a list of values with keys to create a new DataFrame.
district_summary_df = pnds.DataFrame(
          [{"Total Schools": school_count,
          "Total Students": count_studs,
          "Total Budget": ttl_budget,
          "Average Math Score": avg_math_score,
          "Average Reading Score": avg_read_score,
          "% Passing Math": math_passed_perc,
         "% Passing Reading": reading_passed_perc,
        "% Overall Passing": both_passed_perc}])
district_summary_df
```

```python
# formatting total students with , separator  
# declare the change by calling the coloumn itself


district_summary_df['Total Students'] = district_summary_df['Total Students'].map('{:,}'.format)

district_summary_df['Total Students']
```

```python
#formatting budget with dolla sign and , sep

district_summary_df['Total Budget'] = district_summary_df['Total Budget'].map('${:,.2f}'.format)

district_summary_df['Total Budget']
```

```python
# formatting the rest of the coloumns

district_summary_df['Average Math Score'] = district_summary_df['Average Math Score'].map('{:.1f}'.format)

district_summary_df['Average Reading Score'] = district_summary_df['Average Reading Score'].map('{:.1f}'.format)

district_summary_df['% Passing Math'] = district_summary_df['% Passing Math'].map('{:.0f}%'.format)

district_summary_df['% Passing Reading'] = district_summary_df['% Passing Reading'].map('{:.0f}%'.format)

district_summary_df['% Overall Passing'] = district_summary_df['% Overall Passing'].map('{:.0f}%'.format)

district_summary_df
```
